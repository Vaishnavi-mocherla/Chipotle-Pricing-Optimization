```python
# Define the values to be removed
values_to_remove = ['WHITE', 'YELLOW','PEARL TYPE','SETS']

# Remove rows where 'car_type' is in the list of values to remove
df1 = df[~df['Variety'].isin(values_to_remove)]
```

```python
df1['total_price'] = df1.apply(lambda row: row[['Low Price', 'High Price']].mean(), axis=1)
```

```python
# NORMALIZE THE PACKAGE COLUMN BY CONVERTING ALL OF THE ROWS TO LB SCALE , CREATE A NEW COLUMN TO STORE THE  VALUES IN LB SCALE 

# Define a mapping dictionary for normalization
package_mapping = {
    '25 lb sacks': 25,
    '10 lb sacks': 10,  
    'master container 20 2-lb mesh sacks': 40,
    'master container 24 2-lb mesh sacks': 48,
    '50 lb sacks': 50,  
    'master container 5 10-lb mesh sack': 50,  
    '40 lb cartons': 530, # 1 carton is 13 pounds 
    'cartons 12 2-lb mesh sacks': 156, # 12*13 
    'master container 16 3-lb film bags': 2304, # 1 film bag is 48lbs
    '25 lb cartons' : 325,
    'master container 5 10-lb film bags': 2400,
    'nan': 0,
    'master container 10 5-lb mesh sacks': 50,
    'master container 18 2-lb mesh sacks' : 36,
}

# Map the values in the 'Package' column to numeric values
df1['Package Size in pound'] = df['Package'].map(package_mapping)

# Display the preprocessed data
df1
```

```python
df1['Price per Pound'] = df1['total_price'] / df1['Package Size in pound']
```
