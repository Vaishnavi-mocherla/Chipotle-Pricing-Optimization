```python
df=pd.read_csv('Cilantro.csv') 
columns_to_drop = ['Sub Variety', 'Grade', 'Mostly Low','Mostly High','Origin','Origin District', 'Item Size', 'Color',
        'Environment', 'Unit of Sale', 'Quality', 'Condition', 'Appearance',
        'Storage', 'Crop', 'Repack', 'Trans Mode', 'Offerings', 'Market Tone','Price Comment', 'Comments']
df = df.drop(columns=columns_to_drop) # DROP UNWANTED COLUMNS
df = df.dropna(subset=['Package']) # DROP ROWS WITH NULL VALUES IN PACKAGE COLUMN

df['Average Price'] = (df['Low Price'] + df['High Price']) / 2 # FIND THE AVERAGE PRICE USING HIGH AND LOW PRICE
df['Type'] = df['Type'].fillna('Inorganic') #FILL NULL ROWS IN TYPE WITH INORGANIC 
df['Variety'] = df['Variety'].fillna('WITHOUT ROOT') #FILL NULL ROWS IN VARIETY WITH WITHOUT ROOT
df = df.dropna(subset=['Average Price']) # DROP ROWS WITH NULL VALUES IN AVERAGE PRICE 
df
```

```python
distinct_values = df['Package'].unique()

# Display the distinct values
distinct_values
```

```python
# NORMALIZE THE PACKAGE COLUMN BY CONVERTING ALL OF THE ROWS TO LB SCALE , CREATE A NEW COLUMN TO STORE THE  VALUES IN LB SCALE 

# Define a mapping dictionary for normalization
package_mapping = {
    'cartons bunched': 20,
    'crates bunched': 20,  
    'containers approx 18 lbs': 18,
    'cartons': 20,
    '1 1/9 bushel crates bunched':1.0 / 9 * 32.57,  # Assuming 1 bushel is 32.57 lbs
    '1 kg cartons bunched': 20 * 2.20462,  # Convert kg to lb
    '4/5 bushel crates bunched': 4/5 * 32.57,  # Assuming 1 bushel is 40 lbs
    'cartons loose': 20,
    'bushel crates bunched': 32.57,  # Assuming 1 bushel is 40 lbs
    '1/2 bruce crates bunched': 1/2 * 32.57,  # Assuming 1 bushel is 40 lbs
}

# Map the values in the 'Package' column to numeric values
df['Package Size in pound'] = df['Package'].map(package_mapping)

# Display the preprocessed data
df

```

```python
df.count()
```

```python
df['Price per Pound'] = df['Average Price'] / df['Package Size in pound']
```
