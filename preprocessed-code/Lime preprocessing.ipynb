```python
df=pd.read_csv('Limes.csv') 
columns_to_drop = ['Item Size','Origin','Mostly Low','Mostly High','Sub Variety','Grade','Condition', 'Appearance', 'Storage','Crop', 'Repack', 'Trans Mode', 'Offerings', 'Market Tone', 'Price Comment', 'Comments','Origin District','Color','Environment','Unit of Sale','Quality']
df = df.drop(columns=columns_to_drop) # DROP UNWANTED COLUMNS
df = df.dropna(subset=['Package']) # DROP ROWS WITH NULL VALUES IN PACKAGE COLUMN

df['Average Price'] = (df['Low Price'] + df['High Price']) / 2 # FIND THE AVERAGE PRICE USING HIGH AND LOW PRICE
df['Type'] = df['Type'].fillna('Inorganic') #FILL NULL ROWS IN TYPE WITH INORGANIC 
df['Variety'] = df['Variety'].fillna('SEED TYPE') #FILL NULL ROWS IN VARIETY WITH SEED TYPE
df = df.dropna(subset=['Average Price']) # DROP ROWS WITH NULL VALUES IN AVERAGE PRICE 
df
```

```python
distinct_values = df['Package'].unique()

# Display the distinct values
distinct_values
```

```python
# NORMALIZE THE PACKAGE COLUMN BY CONVERTING ALL OF THE ROWS TO LB SCALE , CREATE A NEW COLUMN TO STORE THE  VALUES IN LB SCALE 
package_mapping = {
    '10 lb cartons': 10,
    '40 lb cartons': 40,
    'containers 17 2-lb bags': 17 * 2.20462,  # Convert kg to lb
    'flats 10 1-lb mesh bags': 10,
    '40 lb containers': 40,
    '35-36 lb cartons': 36,
    '38 lb cartons': 38,
    '4 kg containers': 4 * 2.20462,  # Convert kg to lb
    'reusable plastic cntrs 18 2-lb mesh bags': 18 * 2.20462,  # Convert kg to lb
    '15 kg cartons': 15 * 2.20462,  # Convert kg to lb
    '20 kg cartons': 20 * 2.20462,  # Convert kg to lb
    '18 kg cartons': 18 * 2.20462,  # Convert kg to lb
    '16 kg cartons': 16 * 2.20462,  # Convert kg to lb
    '4/5 bushel cartons': 5 * 2.20462,  # Convert kg to lb
    'cartons 18 2-lb bags': 18 * 2.20462,  # Convert kg to lb
}

# Map the values in the 'Package' column to numeric values
df['Package Size in pound'] = df['Package'].map(package_mapping)

# Display the preprocessed data
df

```

```python
df['Price per Pound'] = df['Average Price'] / df['Package Size in pound']
```
