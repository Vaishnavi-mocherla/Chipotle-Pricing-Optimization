```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Lemons.csv', low_memory=False)
```

```python
df.head()
```

```python
columns_to_drop = ['Origin','Mostly Low','Mostly High','Variety','Sub Variety','Grade','Condition', 'Appearance', 'Storage','Crop', 'Repack', 'Trans Mode', 'Offerings', 'Market Tone', 'Price Comment', 'Comments','Origin District','Color','Environment','Unit of Sale','Quality','Item Size']
df = df.drop(columns=columns_to_drop, errors='ignore') # DROP UNWANTED COLUMNS
df = df.dropna(subset=['Package']) # DROP ROWS WITH NULL VALUES IN PACKAGE COLUMN
```

```python
df
```

```python
df['Average Price'] = (df['Low Price'] + df['High Price']) / 2 # FIND THE AVERAGE PRICE USING HIGH AND LOW PRICE
df['Type'] = df['Type'].fillna('Inorganic') #FILL NULL ROWS IN TYPE WITH INORGANIC 
df = df.dropna(subset=['Average Price']) # DROP ROWS WITH NULL VALUES IN AVERAGE PRICE 
df
```

```python
distinct_values = df['Package'].unique()

# Display the distinct values
distinct_values
```

```python
df
```

```python

df = df.copy()

# Define the package mapping
package_mapping = {
    '17 kg cartons': 17 * 2.20462,  # Convert kg to lb
    '7/10 bushel cartons': (7/10) * 2.20462,  # Convert bushel to lb
    'cartons 18 2-lb film bags': 18 * 2.20462,  # Convert kg to lb
    'cartons': 38,  # Set the weight of 1 carton in pounds
    '15 kg containers': 15 * 2.20462,  # Convert kg to lb
    '15 kg cartons': 15 * 2.20462,  # Convert kg to lb
    '18 kg containers': 18 * 2.20462,  # Convert kg to lb
    '4/5 bushel cartons': (4/5) * 2.20462,  # Convert bushel to lb
    'cartons 12 3-lb sacks': 12 * 2.20462,  # Convert lb to lb (no conversion needed)
    '10 kg containers': 10 * 2.20462,  # Convert kg to lb
    '11-12 kg containers': 11 * 2.20462,  # Convert kg to lb (assuming 11 kg)
}

# Map the values in the 'Package' column to numeric values
df['Package Size in pound'] = df['Package'].map(package_mapping)

# Display the preprocessed data
df
```

```python
df.count()
```

```python
# Calculate 'Price per Pound' and assign it to the DataFrame
df = df.assign(**{'Price per Pound': df['Average Price'] / df['Package Size in pound']})
```
